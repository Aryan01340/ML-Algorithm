```python
df=df.drop(columns=['Patient_ID','Age'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
lab_enc=LabelEncoder()
```

```python
df1=lab_enc.fit_transform(df['Gender'])
df['Gender']=df1
df1=lab_enc.fit_transform(df['Country_Region'])
df['Country_Region']=df1
df1=lab_enc.fit_transform(df['Year'])
df['Year']=df1
df1=lab_enc.fit_transform(df['Cancer_Stage'])
df['Cancer_Stage']=df1
df1=lab_enc.fit_transform(df['Cancer_Type'])
df['Cancer_Type']=df1
```

```python
x=df.drop('Treatment_Cost_USD',axis=1)
```

```python
y=df[['Treatment_Cost_USD']]
```

```python
scaler=StandardScaler()
```

```python
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size = 0.3 , random_state = 42)
```

```python
ln = LinearRegression()
```

```python
ln.fit(x_scaled,y)
```

```python
y_predict = ln.predict(x_train)
y_predict
```

```python
x_predict = ln.predict(x_test)
x_predict
```

```python
r2_score(y_train,y_predict)
```

```python
r2_score(y_test,x_predict)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn=KNeighborsRegressor()
```

```python
knn.fit(x_scaled,y)
```

```python
x_pred=knn.predict(x_test)
x_pred
```

```python
y_pred=knn.predict(x_train)
y_pred
```

```python
r2_score(y_train,y_pred)
```

```python
r2_score(y_test,x_pred)
```
